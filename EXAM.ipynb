```python
import numpy as np
import matplotlib.pyplot as plt
price=np.array([20,23,34,24,26,29,32])
per_day=np.array([22,25,39,27,29,31,20])
colors=np.array(["red","green","blue","orange","pink","yellow","black"])
#plt.figure(figuresize=(12,10))
plt.scatter(price,per_day,s=200,c=colors)
plt.xlabel="price"
plt.ylabel="per day"
plt.title("scatterplot of two numerical variables")
plt.show()
```
